```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from skimage import filters

from google.colab import drive
drive.mount('/content/drive')

im_og = cv.imread('/content/drive/MyDrive/Image_Processing/rice.png')
im_og = cv.cvtColor(im_og, cv.COLOR_BGR2GRAY)

# Apply multi-level Otsu thresholding
thresholds = filters.threshold_multiotsu(im_og)
im_2 = np.digitize(im_og, bins=thresholds)

fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(im_og, cmap='gray')
ax[0].set_title('Original')
ax[1].imshow(im_2, cmap='gray')
ax[1].set_title('Segmented Image using Otsu\'s Method')

plt.show()

```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

im_og = cv.imread('/content/drive/MyDrive/Image_Processing/se.png')
im_og = cv.cvtColor(im_og, cv.COLOR_BGR2GRAY)

kernel = cv.getStructuringElement(cv.MORPH_RECT, (3,3))
im_2 = cv.morphologyEx(im_og, cv.MORPH_CLOSE, kernel)

fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(im_og, cmap='gray')
ax[0].set_title('Original')
ax[1].imshow(im_2, cmap='gray')
ax[1].set_title('Output')

plt.show()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

im_og = cv.imread('/content/drive/MyDrive/Image_Processing/morph.png')
im_og = cv.cvtColor(im_og, cv.COLOR_BGR2GRAY)

_, im_1 = cv.threshold(im_og, 128, 255, cv.THRESH_BINARY)

num_labels, labels = cv.connectedComponents(im_1)

fig, ax = plt.subplots(1, 3, figsize=(10, 10))
ax[0].imshow(im_og, cmap='gray')
ax[0].set_title('Original')
ax[1].imshow(im_1, cmap='gray')
ax[1].set_title('Output')
ax[2].imshow(labels, cmap='jet')
ax[2].set_title('Connected Components'.format(num_labels))

plt.show()
print('Number of connected components:', num_labels - 1)

```
